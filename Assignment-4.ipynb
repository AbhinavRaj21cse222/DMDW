## Correlation Co-efficient realation and Co varience analysis 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import math
import statistics as st
% matplotlib in line
```

```python
std=pd.read_csv('student-mat.csv')
std.head(10)
```

```python
std.tail(20)
```

```python
std.corr()
```

```python
columns_of_interest2 = [ 'studytime','freetime']
```

```python
subset_data = std[columns_of_interest2]
```

```python
correlation_matrix = subset_data.corr()
```

```python
covariance_matrix = subset_data.cov()
```

```python
print("Correlation Matrix:")
print(correlation_matrix)
```

```python
sns.heatmap(correlation_matrix,cmap ="YlGnBu")
plt.show()
```

```python
print("\nCovariance Matrix:")
print(covariance_matrix)
```

```python
sns.heatmap(covariance_matrix,cmap ="coolwarm")
plt.show()
```
