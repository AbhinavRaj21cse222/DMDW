```python
cols = A3.columns
```

```python
num_cols = A3._get_numeric_data().columns
```

```python
cat_cols = list(set(cols) - set(num_cols))
```

```python
binary_cols = []
nominal_cols = []
for col in cat_cols:
    if A3[col].nunique() == 2:
        binary_cols.append(col)
    else:
        nominal_cols.append(col)

print('Numerical Attributes:', num_cols)
print('Binary Attributes:', binary_cols)
print('Nominal Attributes:', nominal_cols)
```
